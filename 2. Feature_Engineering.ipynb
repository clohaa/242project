```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file_path1 = 'drive/MyDrive/242 Project/Clean dataset/clean_covid_data.csv'
with open(file_path1, 'r') as f1:
  df = pd.read_csv(f1)
df.drop(['Unnamed: 0'], axis = 1, inplace = True)

df.head()
```

```python
# calculate correlation matrix
corr = df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
# drop highly correlated independent variables

drop_columns = ['total_vaccinations_per_hundred','people_vaccinated_per_hundred','median_age','aged_70_older','life_expectancy','human_development_index']
df.drop(labels=drop_columns, axis = 1, inplace = True)
```

```python
# calculate correlation matrix
corr = df.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```
